```python
# Step 0: Import libraries
import mysql.connector
import pandas as pd

# Step 1: Connect to MySQL
conn = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="bike",
    database="dataspark"
)
cursor = conn.cursor()
```

```python
#1️ Total Sales by Product
query1 = """
SELECT p.ProductName, SUM(s.Quantity) AS Total_Sales
FROM sales s
JOIN products p ON s.ProductKey = p.ProductKey
GROUP BY p.ProductName;
"""

df_total_sales = pd.read_sql(query1, conn)
df_total_sales.head()  # Shows first 5 rows
```

```python
#2️ Most Sold Product
query2 = """
SELECT p.ProductName, SUM(s.Quantity) AS Total_Sales
FROM sales s
JOIN products p ON s.ProductKey = p.ProductKey
GROUP BY p.ProductName
ORDER BY Total_Sales DESC
LIMIT 10;
"""

df_most_sold = pd.read_sql(query2, conn)
df_most_sold
```

```python
#3 Total Quantity by Store and Product

query3 = """
SELECT st.StoreKey, p.ProductName, SUM(s.Quantity) AS Total_Quantity
FROM sales s
JOIN products p ON s.ProductKey = p.ProductKey
JOIN store st ON s.StoreKey = st.StoreKey
GROUP BY st.StoreKey, p.ProductName;
"""

df_quantity_store_product = pd.read_sql(query3, conn)
df_quantity_store_product
```

```python
#4️ Exchange Rate for Specific Date
query4 = """
SELECT Exchange
FROM exchange_rates
WHERE Date = '1/1/2015' AND Currency = 'EUR';
"""

df_exchange_rate = pd.read_sql(query4, conn)
df_exchange_rate
```

```python
#5️ Average Order quantity sold Size by Store
query5 = """
SELECT st.StoreKey, AVG(s.Quantity) AS Average_Order_Size
FROM sales s
JOIN store st ON s.StoreKey = st.StoreKey
GROUP BY st.StoreKey;
"""

df_avg_order_size = pd.read_sql(query5, conn)
df_avg_order_size.head()
```

```python
#6️Total Sales by Currency
query6 = """
SELECT s.CurrencyCode, SUM(s.Quantity) AS Total_Sales
FROM sales s
GROUP BY s.CurrencyCode;
"""

df_sales_currency = pd.read_sql(query6, conn)
df_sales_currency
```

```python
# 7 Sales for Specific Product by Year
query7 = """
SELECT EXTRACT(YEAR FROM s.OrderDate) AS Year, SUM(s.Quantity) AS Total_Sales
FROM sales s
JOIN products p ON s.ProductKey = p.ProductKey
WHERE p.ProductName = 'Contoso 512MB MP3 Player E51 Silver'
GROUP BY Year;
"""

df_sales_year = pd.read_sql(query7, conn)
df_sales_year
```

```python
#8️ Total Quantity Sold per Month
query8 = """
SELECT EXTRACT(MONTH FROM s.OrderDate) AS Month, SUM(s.Quantity) AS Total_Quantity
FROM sales s
GROUP BY Month;
"""

df_quantity_month = pd.read_sql(query8, conn)
df_quantity_month
```

```python
# 9 Customers Who Purchased in More Than 5 Stores
query9 = """
SELECT c.Name, COUNT(DISTINCT s.StoreKey) AS Store_Count
FROM sales s
JOIN customer c ON s.CustomerKey = c.CustomerKey
GROUP BY c.Name
HAVING COUNT(DISTINCT s.StoreKey) > 5;
"""

df_customers_5stores = pd.read_sql(query9, conn)
df_customers_5stores.head()
```

```python
#10 Products Not Sold
query10 = """
SELECT p.ProductName
FROM products p
LEFT JOIN sales s ON p.ProductKey = s.ProductKey
WHERE s.OrderNumber IS NULL;
"""

df_products_not_sold = pd.read_sql(query10, conn)
df_products_not_sold.head()
```
